```python
import pandas as pd
fp = "/content/drive/Shareddrives/IN+/LEIS Colaborative Project/Mapbox Movement/268.csv"
df = pd.read_csv(fp,header=None)
```

```python
df.columns = ["Radio", "MCC", "MNC", "LAC", "CID", "Unit", "Longitude", "Latitude",
              "Range", "Samples", "Changeable", "Created", "Updated", "AverageSignal"]
df
```

```python
import numpy as np
# Ensure no division by zero for Range
df["Range_km"] = df["Range"] / 1000  # Convert range to kilometers
df["Area_km2"] = np.pi * (df["Range_km"] ** 2)  # Coverage area of the tower
df["Density"] = df["Samples"] / df["Area_km2"]  # Samples per km²

# Replace infinities or NaN (if any range is 0)
df["Density"] = df["Density"].replace([np.inf, -np.inf], np.nan).fillna(0)
```

```python
import pandas as pd
#Movement Distribution
url = 'https://raw.githubusercontent.com/INmais/LEIS-2024-collaborative-project/refs/heads/main/Mobility/1922039342088483_2024-11-12.csv'
meta = pd.read_csv(url)
```

```python
meta
```

```python
filtered_meta = meta[meta['country'] == 'PRT']
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


filtered_meta['ds'] = pd.to_datetime(filtered_meta['ds'])


g = sns.FacetGrid(filtered_meta, col="gadm_name", col_wrap=4, height=4, aspect=1, sharey=True)
g.map(sns.lineplot, "ds", "distance_category_ping_fraction", "home_to_ping_distance_category", marker="o")


g.add_legend(title="Distance from Home")
g.set_axis_labels("Date", "Ping Fraction")
g.set_titles("{col_name}")
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Movement Distribution of Users for Each Region")
plt.xticks(rotation=90)


#plt.show()
```

```python
import pandas as pd
import plotly.express as px


filtered_meta['ds'] = pd.to_datetime(filtered_meta['ds'])

# Filter the first 10 cities
top_10_cities = filtered_meta['gadm_name'].drop_duplicates().head(10)
filtered_meta_top10 = filtered_meta[filtered_meta['gadm_name'].isin(top_10_cities)]

# Plot using Plotly
fig = px.line(
    filtered_meta_top10,
    x="ds",
    y="distance_category_ping_fraction",
    color="home_to_ping_distance_category",
    facet_col="gadm_name",
    facet_col_wrap=5,
    markers=True,
    title="Movement Distribution of Users for Top 10 Regions",
    labels={
        "ds": "Date",
        "distance_category_ping_fraction": "Ping Fraction",
        "home_to_ping_distance_category": "Distance from Home"
    }
)

fig.update_layout(
    legend_title="Distance from Home",
    #xaxis_title="Date",
    yaxis_title="Ping Fraction",
    title_x=0.5,
    title_y=0.95
)

fig.update_xaxes(tickangle=90)

fig.show()
```
